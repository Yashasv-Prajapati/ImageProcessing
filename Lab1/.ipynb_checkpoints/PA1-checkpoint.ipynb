```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading image
img = cv.imread("images/reference.jpg")
plt.imshow(img)
```

```python
height = img.shape[0]
width = img.shape[1]

height = int(height*0.25)
width = int(width*0.25)
NNimg = cv.resize(img, (width, height), interpolation = cv.INTER_NEAREST)
plt.imshow(NNimg)
```

```python

centerX, centerY = width//2, height//2
print(width, height)
print(centerX, centerY)
M = cv.getRotationMatrix2D((centerX, centerY), 0, 1)
rotatedImg = cv.warpAffine(img, M, (width, height))
plt.imshow(rotatedImg)
```
